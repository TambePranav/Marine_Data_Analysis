```python
import pandas as pd
```

```python
data=pd.read_csv('./graph_data_cdom.csv')
```

```python
data1 = pd.DataFrame()
new_col=[]
i=0
for column in data:
    i+=1
    if(i%3==0):
        data1 = pd.concat([data1,data[column]], axis=1)
    elif(i%3==1):
        new_col.append(data[column][1])
    
```

```python
data1=data1.T
```

```python
print(data1)
```

```python
print(new_col)
print(f"\nlength is {len(new_col)}")
```

```python
data1.insert( 0,'cdom',new_col,True)
```

```python
data1.to_csv('Transposed_graph_data_cdom.csv', index=False)
```
